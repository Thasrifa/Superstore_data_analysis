```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Data Loading and Initial Overview
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
df.head(10)
```

```python
df.shape
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
df.dtypes
df.info()
df.describe()
```

**Data Pre-Proceccing**

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
#Correcting Datatypes
import pandas as pd

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce',dayfirst=True)
df['Ship Date']  = pd.to_datetime(df['Ship Date'], errors='coerce', dayfirst=True)
df['Sales']=pd.to_numeric(df['Sales'],errors='coerce')
df['Quantity']=pd.to_numeric(df['Quantity'],errors='coerce')
df['Discount']=pd.to_numeric(df['Discount'],errors='coerce')
df['Profit']=pd.to_numeric(df['Profit'],errors='coerce')
df.dtypes
```

```python
#Derived Columns
df["ProfitRatio"]=df["Profit"]/df["Sales"]
df["ShippingTime"]=df["Ship Date"]-df["Order Date"]
df["OrderMonth"]=df["Order Date"].dt.month
df["OrderYear"]=df["Order Date"].dt.year
df['OrderYear'] = df['Order Date'].dt.year.astype('Int64')
df['OrderMonth'] = df['Order Date'].dt.month.astype('Int64')
df.head(6)
```

```python
df['Order Date'].isna().sum()
```

**Filtering&Aggregating**

```python
# Filter 2016 Data
df_2016 = df[df['OrderYear'] == 2016]
print("Rows for 2016:", df_2016.shape[0])
#Average Discount By Region
avg_discount_region = df.groupby('Region')['Discount'].mean()
print("Average Discount by Region:", avg_discount_region)
# Total sales and profit by category
sales_by_category = df.groupby('Category')[['Sales', 'Profit']].sum()
print("Sales and Profit by Category:", sales_by_category)
```

**Visualizations**

```python
#Distribution Of Ship Modes
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Ship Mode')
plt.title('Distribution of Ship Modes')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```

```python
#Number Of Orders By Category
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category')
plt.title('Number of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution Of Sales
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=20, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Profit Distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Profit', bins=20, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
#Sales By Category
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Sales')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
#Profit By Region
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Region', y='Profit')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

```python
#Sales Vs  Profit Scatter
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit',s=60)
plt.title('Sales vs Profit Scatter')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
#Profit Distribution By Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit')
plt.title('Profit Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
#Discount Distribution
plt.hist(df['Discount'], bins=20, color='#66C2A5', edgecolor='black')
plt.title('Distribution of Discounts', fontsize=14, fontweight='bold')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numerical Features', fontsize=13, fontweight='bold')
plt.show()
```

```python
#Orders by Ship Mode
plt.figure(figsize=(6,4))
sns.countplot(x='Ship Mode', data=df, palette='Set1')
plt.title('Number of Orders by Ship Mode', fontsize=14, fontweight='bold')
plt.xlabel('Ship Mode')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Sales Distribution By Region
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(5,5))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set1'))
plt.title('Sales Distribution by Region')
plt.show()
```

```python
#Sales VS Quantity
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Sales', y='Quantity', hue='Category', palette='Set1')
plt.title('Sales vs Quantity', fontsize=14, fontweight='bold')

```

```python
#Profit By Segment
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Segment', y='Profit', palette='Set1')
plt.title('Profit Distribution by Segment', fontsize=14, fontweight='bold')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()
```
