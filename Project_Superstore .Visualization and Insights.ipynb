```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Data Loading and Initial Overview
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
df.head(10)
```

```python
df.shape
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
df.dtypes
df.info()
df.describe()
```

**Data Pre-Proceccing**

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
#Correcting Datatypes
import pandas as pd

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce',dayfirst=True)
df['Ship Date']  = pd.to_datetime(df['Ship Date'], errors='coerce', dayfirst=True)
df['Sales']=pd.to_numeric(df['Sales'],errors='coerce')
df['Quantity']=pd.to_numeric(df['Quantity'],errors='coerce')
df['Discount']=pd.to_numeric(df['Discount'],errors='coerce')
df['Profit']=pd.to_numeric(df['Profit'],errors='coerce')
df.dtypes
```

```python
#Derived Columns
df["ProfitRatio"]=df["Profit"]/df["Sales"]
df["ShippingTime"]=df["Ship Date"]-df["Order Date"]
df["OrderMonth"]=df["Order Date"].dt.month
df["OrderYear"]=df["Order Date"].dt.year
df['OrderYear'] = df['Order Date'].dt.year.astype('Int64')
df['OrderMonth'] = df['Order Date'].dt.month.astype('Int64')
df['Month'] = df['Order Date'].dt.month_name()
df.head(6)
```

```python
df['Order Date'].isna().sum()
```

**Filtering&Aggregating**

```python
# Filter 2016 Data
df_2016 = df[df['OrderYear'] == 2016]
print("Rows for 2016:", df_2016.shape[0])
#Average Discount By Region
avg_discount_region = df.groupby('Region')['Discount'].mean()
print("Average Discount by Region:", avg_discount_region)
# Total sales and profit by category
sales_by_category = df.groupby('Category')[['Sales', 'Profit']].sum()
print("Sales and Profit by Category:", sales_by_category)

```

**Overall Insights**

**1.Sales vs. Profit**

 * While Sales are generally high across multiple regions, Profit margins are inconsistent.


 * Some categories (especially Furniture and Technology) generate high revenue, but not always high profit due to high discounts and shipping costs.

**2.Category Perfomance**

 * **Office Supplies**:Contributes the highest profit margin due to low shipping and steady demand.
 * **Technology**:Drives strong sales volume but has thin margins due to discounting.
 * **Furniture**:Often unprofitable — particularly in “Tables” and “Bookcases” sub-categories.

**3.Regional Insights**

 * **West Region**:Highest overall sales and profit.
 * **Central Region**:Moderate sales but low profit margins.
 * **South Region**:Shows potential for growth but often impacted by shipping costs.

**4.Customer Segments**

 * **Consumer Segment** dominates in both sales and profit.
 * **Corporate Segment** maintains stable, repeat orders.
 * **Home Office** customers purchase smaller but higher-margin items.

**5.Shipping Mode Impact**
  
 * **Same Day** shipping yields highest profit per order (less discounting).
 * **Standard Class** accounts for most orders but offers lower margins due to bulk and discounted pricing.

**6.Discount vs. Profit Correlation**

 * Strong **negative correlation** between discount rate and profit — as discounts increase, profits drop.
 * High discounting may attract buyers but hurts profitability.

**7.Top Performing States**

 * **California** and **New York** lead in both sales and profit.
 * **Texas**   show good sales but lower profit margins.

**8.Loss-Making Areas**

 * Certain sub-categories like Tables, Bookcases, and Supplies often show negative profit margins.
 * **High shipping cost  and large discounts** are key contributors to these losses.

**9.Yearly Trends**

* Sales and profit show steady growth over years — indicating overall business expansion.
* Seasonal peaks are observed  during **Q4 (October–December)** , aligning with holiday shopping


























**Actionable Insights**

 * **Optimize discounts**: Limit heavy discounting in unprofitable sub-categories.
 * **Focus marketing** on high-margin categories like Office Supplies and Accessories.
 * **Improve shipping efficiency** in Central and South regions to reduce losses.
 * **Boost corporate partnerships**  for recurring bulk orders.
 * **Analyze loss-making states/sub-categories** for pricing or supply chain improvements.
