```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Data Loading and Initial Overview
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
df.head(10)
```

```python
df.shape
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
df.dtypes
df.info()
df.describe()
```

```python
#Data Pre-Processing
df.isnull().sum()
df.duplicated().sum()
```

```python
#Correcting Datatypes
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
df['Sales']=pd.to_numeric(df['Sales'],errors='coerce')
df['Quantity']=pd.to_numeric(df['Quantity'],errors='coerce')
df['Discount']=pd.to_numeric(df['Discount'],errors='coerce')
df['Profit']=pd.to_numeric(df['Profit'],errors='coerce')
df.dtypes
```

```python
#Derived Columns
df["ProfitRatio"]=df["Profit"]/df["Sales"]
df["ShippingTime"]=df["Ship Date"]-df["Order Date"]
df["OrderMonth"]=df["Order Date"].dt.month
df["OrderYear"]=df["Order Date"].dt.year
df['OrderYear'] = df['Order Date'].dt.year.astype('Int64')
df['OrderMonth'] = df['Order Date'].dt.month.astype('Int64')
df.head(6)
```

```python
df['Order Date'].isna().sum()
```

```python
#Filtering&Aggregating
# Filter 2016 Data
df_2016 = df[df['OrderYear'] == 2016]
print("Rows for 2016:", df_2016.shape[0])
#Average Discount By Regon
avg_discount_region = df.groupby('Region')['Discount'].mean()
print("Average Discount by Region:", avg_discount_region)
# Total sales and profit by category
sales_by_category = df.groupby('Category')[['Sales', 'Profit']].sum()
print("Sales and Profit by Category:", sales_by_category)
```
